```python
import os
from collections import defaultdict

# Label fayllarÄ±nÄ±n yolu
label_folder = "/content/ppe_dataset/data/labels/train"  # Ã–z yolunla É™vÉ™z et

# Class sayÄŸacÄ±
class_counts = defaultdict(int)

# Label fayllarÄ±nÄ± oxu
for file_name in os.listdir(label_folder):
    if file_name.endswith(".txt"):
        with open(os.path.join(label_folder, file_name), "r") as f:
            for line in f:
                if line.strip():  # boÅŸ sÉ™trlÉ™ri keÃ§
                    class_id = int(line.strip().split()[0])
                    class_counts[class_id] += 1

# NÉ™ticÉ™lÉ™ri Ã§ap et
print("Class tÉ™krar saylarÄ±:")
for class_id in sorted(class_counts):
    print(f"Class {class_id}: {class_counts[class_id]} dÉ™fÉ™")
```
